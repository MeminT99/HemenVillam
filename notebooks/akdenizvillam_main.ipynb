```python
import bs4
import requests
import pandas as pd
import re
```

```python
urls = []

for page_number in range(1, 188):
    URL = f"https://www.akdenizvillam.com/tum-kiralik-villalar/page/{page_number}"
    response = requests.get(URL)
    response.encoding = "utf-8" 
    html_parsed = bs4.BeautifulSoup(response.text, "html.parser")
    
    urls.extend(set([villa_elem["href"] for villa_elem in html_parsed.select(".px-4.pb-3 a")]))
        
    if page_number % 10 == 0:
        print(f"{page_number} pages have been checked!")
        
print(f"\n---OPERATION HAS BEEN COMPLETED!---")
print(f"{page_number} pages have been checked!")
print(f"Total Villa URL: {len(urls)}")
```

```python
addresses = []
names = []
ids = []
guests = []
bedrooms = []
bathrooms = []
descriptions = []
latitudes = []
longitudes = []
photos = []
average_prices = []
q1_prices = []
q2_prices = []
q3_prices = []
q4_prices = []

URL_ERRORS = []
villa_number = 0
for URL in urls:
    try:
        response = requests.get(URL)
        response.encoding = "utf-8" 
        html_parsed = bs4.BeautifulSoup(response.text, "html.parser")
        villa_number += 1
    except:
        print(f"{URL} : URL ERROR!")
        URL_ERRORS = []
        
    try:
        address = html_parsed.select_one(".d-block.d-md-flex.flex-horizontal-center.font-size-14.text-gray-1").getText().strip().split()[0]
    except:
        print(f"{URL} : ADDRESS ERROR!")
        guest = None
        
    try:
        name = html_parsed.select_one(".font-size-23.font-weight-bold.mb-1").getText().strip()
        villa_id = name.replace(" ", "").lower()
    except:
        print(f"{URL} : NAME ERROR!")
        name = None
        villa_id = None
        
    try:
        guest = int(html_parsed.select(".text-gray-1.ozel-gray")[1].getText().strip().split()[0])
    except:
        print(f"{URL} : GUEST NUMBER ERROR!")
        guest = None
        
    try:
        bedroom = int(html_parsed.select(".text-gray-1.ozel-gray")[3].getText().strip().split()[0])
    except:
        print(f"{URL} : BEDROOM NUMBER ERROR!")
        bedroom = None
    
    try:
        bathroom = int(html_parsed.select(".text-gray-1.ozel-gray")[2].getText().strip().split()[0])
    except:
        print(f"{URL} : BATHROOM NUMBER ERROR!")
        bathroom = None
        
    try:
        description = ("\n".join([text_elem.getText().strip() for text_elem in html_parsed.select("#desc_two p")])).strip()
    except:
        print(f"{URL} : DESCRIPTION ERROR!")
        description = None
        
    try:
        loc_elem = str(html_parsed.select(".card.card_ser .btn.btn-white.transition-3d-hover.ml-2.py-2.px-md-5.px-3.shadow-6")[1]["href"])
        latitude, longitude = [float(elem.strip()) for elem in loc_elem.split("destination=")[1].split(",")]
    except:
        print(f"{URL} : LOCATION ERROR!")
        latitude = None
        longitude = None
        
    try:
        villa_photos = tuple(villa_photo["data-src"] for villa_photo in html_parsed.select(".js-fancybox.d-none"))
    except:
        print(f"{URL} : PHOTO ERROR")
        villa_photos = None
        
    try:
        villa_prices = [int(re.sub("[^0-9]", "", price_elem.getText().strip())) for price_elem in html_parsed.select(".ay-haftalik.font-weight-bold.font-size-22.ml-1.try")[::2]]
        avg_price = sum(villa_prices) // len(villa_prices)
        q1_price = avg_price*4 * 0.14
        q2_price = avg_price*4 * 0.32
        q3_price = avg_price*4 * 0.37
        q4_price = avg_price*4 * 0.17
    except:
        print(f"{URL} : PRICE ERROR!")
        avg_price = None
        q1_price = None
        q2_price = None
        q3_price = None
        q4_price = None
    
    addresses.append(address)
    names.append(name)
    ids.append(villa_id)
    guests.append(guest)
    bedrooms.append(bedroom)
    bathrooms.append(bathroom)
    descriptions.append(description)
    latitudes.append(latitude)
    longitudes.append(longitude)
    average_prices.append(avg_price)
    q1_prices.append(q1_price)
    q2_prices.append(q2_price)
    q3_prices.append(q3_price)
    q4_prices.append(q4_price)
    photos.append(villa_photos)   
    
    if villa_number % 75 == 0:
        print(f"{villa_number} villas have been completed!")

print(f"THE OPERATION IS COMPLETED!")
print(f"TOTAL VILLA NUMBER: {villa_number}")
```

```python
data_raw = {
    "ID": ids,
    "Name": names,
    "Address": addresses,
    "Guest Number": guests,
    "Bedroom Number": bedrooms,
    "Bathroom Number": bathrooms,
    "Latitude": latitudes,
    "Longitude": longitudes,
    "Description": descriptions,
    "Photo URLs": photos,
    "Average Price": average_prices,
    "Q1 Price": q1_prices,
    "Q2 Price": q2_prices,
    "Q3 Price": q3_prices,
    "Q4 Price": q4_prices
}

df_raw = pd.DataFrame(data_raw)
df_raw.set_index("ID", inplace=True)
df_raw.to_excel("akdenizvillam_raw.xlsx")

data_info = {
    "ID": ids,
    "Name": names,
    "Address": addresses,
    "Guest Number": guests,
    "Bedroom Number": bedrooms,
    "Bathroom Number": bathrooms,
    "Latitude": latitudes,
    "Longitude": longitudes,
    "Description": descriptions,
    "Photo URLs": photos  
}

df_info = pd.DataFrame(data_info)
df_info.set_index("ID", inplace=True)
df_info.to_excel("akdenizvillam_info.xlsx")

data_financial = {
    "ID": ids,
    "Average Price": average_prices,
    "Q1 Price": q1_prices,
    "Q2 Price": q2_prices,
    "Q3 Price": q3_prices,
    "Q4 Price": q4_prices   
}

df_financial = pd.DataFrame(data_financial)
df_financial.set_index("ID", inplace=True)
df_financial.to_excel("akdenizvillam_financial.xlsx")
```
