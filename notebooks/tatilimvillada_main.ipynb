```python
import bs4
import requests
import pandas as pd
import unicodedata
import re
```

```python
urls = []
names = []
ids = []
addresses = []

for page_number in range(1, 49):
    URL = f"https://www.tatilimvillada.com/villa-filtre?page={page_number}"
    response = requests.get(URL)
    response.encoding = "utf-8" 
    html_parsed = bs4.BeautifulSoup(response.text, "html.parser")
    
    for element in html_parsed.select("div.shop-item .shop-item-body a")[::3]:
        villa_name = element.select_one("h6").getText().strip()
        villa_id = villa_name.lower().replace(" ", "")
        villa_url = "https://www.tatilimvillada.com"+element["href"]
        urls.append(villa_url)
        ids.append(villa_id)
        names.append(villa_name) 
    
    for villa_loc in html_parsed.select(".location"):
        villa_location = villa_loc.getText().split("-")[1].strip().split()[0]
        if villa_location != "Villa":
            addresses.append(villa_location)
        else:
            addresses.append(None)
    
    if page_number % 10 == 0:
        print(f"{page_number} pages have been checked!")
        
print(f"\n---OPERATION HAS BEEN COMPLETED!---")
print(f"{page_number} pages have been checked!")
print(f"Total Villa Name: {len(names)}")
print(f"Total Villa URL: {len(urls)}")
```

```python
average_prices = []
q1_prices = []
q2_prices = []
q3_prices = []
q4_prices = []
guests = []
bedrooms = []
bathrooms = []
photos = []
descriptions = []
latitudes = []
longitudes = []

URL_ERRORS = []
NOT_TRY = []
villa_number = 0
for URL in urls:
    try:
        response = requests.get(URL)
        response.encoding = "utf-8" 
        html_parsed = bs4.BeautifulSoup(response.text, "html.parser")
        villa_number += 1
    except:
        print(f"{URL} : URL ERROR!")
        URL_ERRORS = []
        
    try:
        villa_prices = [int(re.sub("[^0-9]", "", villa_price.getText().split(".")[0].strip())) for villa_price in html_parsed.select("td.text-right.room-price")]
        avg_price = sum(villa_prices) // len(villa_prices)
        q1_price = avg_price*4 * 0.14
        q2_price = avg_price*4 * 0.32
        q3_price = avg_price*4 * 0.37
        q4_price = avg_price*4 * 0.17
    except:
        print(f"{URL} : PRICE ERROR!")
        avg_price = None
        q1_price = None
        q2_price = None
        q3_price = None
        q4_price = None
        
    try:
        if str(html_parsed.select_one("td.text-right.room-price i")["class"][1].split("-")[1].strip()) != "try":
            NOT_TRY.append(URL)
    except:
        print(f"{URL} : FOREIGN PRICE ERROR!")
        NOT_TRY.append(URL)              
        
    try:
        guest = int(html_parsed.select("ul.facility-thumbs li span")[0].getText().strip()[0])
    except:
        print(f"{URL} : GUEST NUMBER ERROR!")
        guest = None
        
    try:
        bedroom = int(html_parsed.select("ul.facility-thumbs li span")[2].getText().strip()[0])
    except:
        print(f"{URL} : BEDROOM NUMBER ERROR!")
        bedroom = None
    
    try:
        bathroom = int(html_parsed.select("ul.facility-thumbs li span")[1].getText().strip()[0])
    except:
        print(f"{URL} : BATHROOM NUMBER ERROR!")
        bathroom = None
        
    try:
        villa_photos = tuple("https://www.tatilimvillada.com"+photo_url["href"] for photo_url in html_parsed.select(".villa-gallery a.hidden"))
    except:
        print(f"{URL} : PHOTO ERROR")
        villa_photos = None
        
    try:
        description = "\n".join([unicodedata.normalize("NFKD" ,txt.getText()) for txt in html_parsed.select_one("#contyazi").select("p")]).strip()
    except:
        try:
            description = "\n".join([unicodedata.normalize("NFKD" ,txt.getText()) for txt in html_parsed.select_one("#_hakkinda").select("p")]).strip()             
        except:
            print(f"{URL} : DESCRIPTION ERROR!")
            description = None
        
    latitude = None
    longitude = None   
    
    guests.append(guest)
    bedrooms.append(bedroom)
    bathrooms.append(bathroom)
    descriptions.append(description)
    latitudes.append(latitude)
    longitudes.append(longitude)
    average_prices.append(avg_price)
    q1_prices.append(q1_price)
    q2_prices.append(q2_price)
    q3_prices.append(q3_price)
    q4_prices.append(q4_price)
    photos.append(villa_photos)   
    
    if villa_number % 50 == 0:
        print(f"{villa_number} villas have been completed!")

print(f"THE OPERATION IS COMPLETED!")
print(f"TOTAL VILLA NUMBER: {villa_number}")
```

```python
data_raw = {
    "ID": ids,
    "Name": names,
    "Address": addresses,
    "Guest Number": guests,
    "Bedroom Number": bedrooms,
    "Bathroom Number": bathrooms,
    "Latitude": latitudes,
    "Longitude": longitudes,
    "Description": descriptions,
    "Photo URLs": photos,
    "Average Price": average_prices,
    "Q1 Price": q1_prices,
    "Q2 Price": q2_prices,
    "Q3 Price": q3_prices,
    "Q4 Price": q4_prices
}

df_raw = pd.DataFrame(data_raw)
df_raw.set_index("ID", inplace=True)
df_raw.to_excel("tatilimvillada_raw.xlsx")

data_info = {
    "ID": ids,
    "Name": names,
    "Address": addresses,
    "Guest Number": guests,
    "Bedroom Number": bedrooms,
    "Bathroom Number": bathrooms,
    "Latitude": latitudes,
    "Longitude": longitudes,
    "Description": descriptions,
    "Photo URLs": photos  
}

df_info = pd.DataFrame(data_info)
df_info.set_index("ID", inplace=True)
df_info.to_excel("tatilimvillada_info.xlsx")

data_financial = {
    "ID": ids,
    "Average Price": average_prices,
    "Q1 Price": q1_prices,
    "Q2 Price": q2_prices,
    "Q3 Price": q3_prices,
    "Q4 Price": q4_prices   
}

df_financial = pd.DataFrame(data_financial)
df_financial.set_index("ID", inplace=True)
df_financial.to_excel("tatilimvillada_financial.xlsx")

data_foreign_curr = {
    "URL": NOT_TRY
}

df_foreign = pd.DataFrame(data_foreign_curr)
df_foreign.to_excel("tatilimvillada_foreign_curr.xlsx")
```
