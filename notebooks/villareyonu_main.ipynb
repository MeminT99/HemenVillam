```python
import requests
import pandas as pd
import bs4
import unicodedata
import re
```

```python
urls = []

for page_number in range(1, 143):
    URL = f"https://www.villareyonu.com/kiralik-villalar?s={page_number}"
    response = requests.get(URL)
    response.encoding = "utf-8"
    html_parsed = bs4.BeautifulSoup(response.text, "html.parser")
    villa_infos = html_parsed.select("div.d-flex.flex-wrap.justify-content-between.pb-1.align-items-center.border-bottom a")

    for villa_info in villa_infos:
        urls.append("https://www.villareyonu.com"+villa_info["href"])
        
    if page_number % 10 == 0:
        print(f"{page_number} pages have been checked!")
        
print(f"\n---OPERATION HAS BEEN COMPLETED!---")
print(f"{page_number} pages have been checked!")
print(f"Total Villa URL: {len(urls)}")
```

```python
names = []
ids = []
addresses = []
photos = []
latitudes = []
longitudes = []
average_prices = []
q1_prices = []
q2_prices = []
q3_prices = []
q4_prices = []
bedrooms = []
guests = []
bathrooms = []
descriptions = []

URL_ERRORS = []
villa_number = 0
for URL in urls:
    try:
        response = requests.get(URL)
        response.encoding = "utf-8" 
        html_parsed = bs4.BeautifulSoup(response.text, "html.parser")
        villa_number += 1
    except:
        print(f"{URL} : URL ERROR!")
        URL_ERRORS.append(URL)
        
    try:
        villa_address = html_parsed.select_one("b.text-dark.order-sm-1.order-2").getText().strip()
    except:
        print(f"{URL} : ADDRESS ERROR!")    
        villa_address = None
        
    try:
        villa_name = html_parsed.select_one("h1.font-weight-bold.mb-0.villaTitle.order-sm-2.order-1.mr-sm-0.mr-3").getText().strip()
        villa_id = villa_name.replace(" ", "").lower()
    except:
        print(f"{URL} : NAME ERROR!")
        villa_name = None
        villa_id = None
        
    try:
        photo = tuple(image["href"] for image in html_parsed.select("div.d-flex.justify-content-center.glryBtns.flex-sm-row.flex-column.align-items-center div.d-none a")) 
    except:
        print(f"{URL} PHOTO ERROR!")
        photo = None
        
    try:
        latitude = float(html_parsed.select_one("div.map.w-100")["data-lat"])
        longitude = float(html_parsed.select_one("div.map.w-100")["data-lng"])
    except:
        print(f"{URL} : LOCATION ERROR!")
        latitude = None
        longitude = None
        
    try:
        villa_prices = [price_text.getText().strip().replace(".", "") for price_text in html_parsed.select("div.d-flex.flex-column.align-items-between.yamukPriceWrap")]
        villa_prices = [int(villa_price.split()[1]) for villa_price in villa_prices]
        avg_price = sum(villa_prices) // len(villa_prices)
        q1_price = avg_price*4 * 0.14
        q2_price = avg_price*4 * 0.32
        q3_price = avg_price*4 * 0.37
        q4_price = avg_price*4 * 0.17
    except:
        avg_price = None
        q1_price = None
        q2_price = None
        q3_price = None
        q4_price = None
        print(f"{URL} : PRICE ERROR!")
        
    try:
        numeric_infos = [int(re.search(r'\d+', num_info.getText()).group()) for num_info in html_parsed.select(".d-flex.justify-content-center.border.rounded-lg.p-sm-3.p-2.my-1.villaKategori")[:3]]

        bedroom = numeric_infos[0]
        guest = numeric_infos[1]
        bathroom = numeric_infos[2]
    except:
        print(f"{URL} : NUMERIC INFO ERROR!")
        guest = None
        bedroom = None
        bathroom = None
        
    try:
        desc_html = html_parsed.select(".overflow-hidden.position-relative.showMoreSection.villaAciklamasi.fs-18 p")
        description = unicodedata.normalize("NFKD", "\n".join([desc.getText().strip() for desc in desc_html]))
    except:
        print(f"{URL} : DESCRIPTION ERROR!")
        description = None
        
    latitudes.append(latitude)
    longitudes.append(longitude)
    photos.append(photo)
    names.append(villa_name)
    ids.append(villa_id)
    addresses.append(villa_address)
    average_prices.append(avg_price)
    q1_prices.append(q1_price)
    q2_prices.append(q2_price)
    q3_prices.append(q3_price)
    q4_prices.append(q4_price)
    descriptions.append(description)
    guests.append(guest)
    bedrooms.append(bedroom)
    bathrooms.append(bathroom)
    
    if villa_number % 80 == 0:
        print(f"{villa_number} villas are done!")
        
print(f"THE OPERATION IS COMPLETED!")
print(f"TOTAL VILLA NUMBER: {villa_number}")       
```

```python
data_raw = {
    "ID": ids,
    "Name": names,
    "Address": addresses,
    "Guest Number": guests,
    "Bedroom Number": bedrooms,
    "Bathroom Number":bathrooms,
    "Latitude": latitudes,
    "Longitude": longitudes,
    "Description": descriptions,
    "Photo URLs": photos,
    "Average Price": average_prices,
    "Q1 Price": q1_prices,
    "Q2 Price": q2_prices,
    "Q3 Price": q3_prices,
    "Q4 Price": q4_prices
}

df_raw = pd.DataFrame(data_raw)
df_raw.set_index("ID", inplace=True)
df_raw.to_excel("villareyonu_raw.xlsx")

data_info = {
    "ID": ids,
    "Name": names,
    "Address": addresses,
    "Guest Number": guests,
    "Bedroom Number": bedrooms,
    "Bathroom Number": bathrooms,
    "Description": descriptions,
    "Photo URLs": photos,
    "Latitude": latitudes,
    "Longitude": longitudes
}

df_info = pd.DataFrame(data_info)
df_info.set_index("ID", inplace=True)
df_info.to_excel("villareyonu_info.xlsx")

data_financial = {
    "ID": ids,
    "Average Price": average_prices,
    "Q1 Price": q1_prices,
    "Q2 Price": q2_prices,
    "Q3 Price": q3_prices,
    "Q4 Price": q4_prices   
}

df_financial = pd.DataFrame(data_financial)
df_financial.set_index("ID", inplace=True)
df_financial.to_excel("villareyonu_financial.xlsx")
```
